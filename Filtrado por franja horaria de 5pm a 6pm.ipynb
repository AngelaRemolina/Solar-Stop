# Selección de datos en el rango de 5pm a 6pm 

## Importar csv

```python
import pandas as pd

df = pd.read_csv (r'C:\Users\Angela\Downloads\radiación solar - SPEC.csv')
df
```

Se pone titulo en cada columna.

```python
df.columns = ['Old Index','Estacion','Ubicacion', 'Proyecto','Longitud','Latitud','Radiacion',"Fecha/Hora"]
df
```

Se elimina la columna del index, para al finalizar la limpieza, reindexar las filas.

```python
df = df.drop(['Old Index'], axis=1)
df
```

Se separa el día de la medición de la hora, y se elimina la columna que las tiene juntas.

```python
df[['Fecha','Hora']] = df['Fecha/Hora'].str.split('T',expand=True)
df = df.drop(columns=['Fecha/Hora'])
df
```

Dado que el formato de la hora varía (a veces inclye segundos, en otras no), se opta por separar en dos columnas las horas y los minutos.

```python
horas = []
minutos= []
for i in df['Hora']:
    
    if(len(i)==2):
        horas.append(i)
        minutos.append("00")
    else: 
        hTemp = i.split(":")
        horas.append(hTemp[0])
        minutos.append(hTemp[1])
    
```

```python
df['Horas'] = horas
df['Minutos'] = minutos
df
```

```python
df['Horas'] = pd.to_numeric(df['Horas'])
df['Minutos'] = pd.to_numeric(df['Minutos'])
df.dtypes
```

```python
df['Valores X'] = ((df['Horas'])*100) + (df['Minutos'])
df
```

Una vez hecho esto, ya solo se toman los valores donde la hora sea igual a 17 y así se tiene el rango de hora de 5:00pm a 5:59pm.

```python
df_5a6 = df.copy() #5pm=17
df_5a6 = df_5a6[(df_5a6.Horas == 17)]
df_5a6
```

Al haber eliminado todos los dems registros, se debe re indexar las filas, de la siguiente manera.

```python
df_5a6 = df_5a6.reset_index(drop=True)
df_5a6
```

Por último se saca el rango de datos ya seleccionado a un archivo CSV.

```python
df_5a6.to_csv('datos-5a6.csv', encoding='utf-8',index=False)
```
